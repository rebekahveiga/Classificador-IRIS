```
# Basic example on how to buld a keras sequential model and train a classifier
# pip3 install tensorflow
# pip3 install keras
# pip3 install numpy
# pip3 install scikit-learn

from __future__ import absolute_import
from __future__ import division
from __future__ import print_function

from sklearn import datasets, metrics
from keras.layers import Dense
from keras.models import Sequential
from keras.utils import to_categorical
import numpy as np
from sklearn.model_selection import train_test_split

# Network model
def build_model():
	# create model
	model = Sequential()
	model.add(Dense(16, input_dim=4, activation='relu'))
	model.add(Dense(16, activation='relu'))
	model.add(Dense(16, activation='relu'))
	model.add(Dense(16, activation='relu'))
	model.add(Dense(3, activation='softmax'))
	# Compile model
	model.compile(loss='categorical_crossentropy', optimizer='sgd', metrics=['accuracy'])
  # print structure
	model.summary()
	return model



# load dataset and convert to onehot encoding
iris = datasets.load_iris()
x = iris.data
y = to_categorical(iris.target)
# print(y)
# split dataset in train / test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)

# build and train model
classifier = build_model()
classifier.fit(x_train, y_train, epochs=500)

# evaluate model on test data
_, score = classifier.evaluate(x_test,y_test)

print("Accuracy on test data: %f" % score)
```
